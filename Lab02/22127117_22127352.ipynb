## **I. Tóm tắt đồ án**

### **1. Giới thiệu đồ án**

[World Development Indicators (WDI)](https://databank.worldbank.org/source/world-development-indicators#) là một nguồn dữ liệu do Ngân hàng Thế giới (World Bank) cung cấp, tập trung vào các chỉ số phát triển toàn cầu. Đây là kho dữ liệu toàn diện với hơn 1.600 chỉ số về kinh tế, xã hội và môi trường, bao gồm thông tin từ hơn 200 quốc gia trên toàn cầu, trải dài qua nhiều năm. Thông qua việc khai thác và phân tích dữ liệu này, đồ án hướng đến mục tiêu trình bày các xu hướng phát triển, khám phá mối quan hệ giữa các chỉ số, và làm rõ các vấn đề nổi bật như phát triển kinh tế, biến đổi khí hậu, ....

### **2. Mục tiêu đồ án**

Trong đồ án nhóm, chúng em sẽ tập trung khai thác và phân tích xu hướng phát triển kinh tế – xã hội của 10 quốc gia phát triển hàng đầu thế giới, đồng thời so sánh với 10 quốc gia kém phát triển nhất. Bên cạnh đó, chúng em cũng sẽ mở rộng phạm vi nghiên cứu sang khu vực Đông Nam Á, nhằm đánh giá tình hình phát triển của các quốc gia lân cận. Qua đó, cung cấp thêm những góc nhìn sâu sắc hơn về môi trường kinh tế – xã hội trong khu vực, để từ ấy có thể đề xuất những chiến lược khả thi giúp Việt Nam nâng cao năng lực cạnh tranh và thúc đẩy tăng trưởng, phát triển bền vững hơn trong khu vực và vươn lên tầm quốc tế.

### **3. Về bộ dữ liệu**

Để có thể đạt được mục đích phân tích, chúng em đã lấy dữ liệu từ [World Development Indicators (WDI)](https://databank.worldbank.org/source/world-development-indicators#), với **63 thuộc tính khác nhau** bao gồm Quốc gia, GDP, GNI, Tăng trưởng GDP, Lạm phát, Cán cân thương mại, Đầu tư, FDI, Tỷ lệ dân số, Lao động, Giáo dục, Y tế, Tuổi thọ, Bất bình đẳng thu nhập, Phát thải CO₂, Năng lượng tái tạo, Tiêu thụ nước, Tài nguyên thiên nhiên, Năm. #### **Xử lý giá trị thiếu**

```python
print("Tỉ lệ % số lượng giá trị thiếu ở mỗi cột")

(df_cleaned.isnull().sum()/df_cleaned.shape[0] * 100).round(3)
```

Trước hết đối với những cột có tỉ lệ % giá trị thiếu trên 80% ta sẽ xóa những cột đó.

```python
threshold = int(0.8 * len(df_cleaned))
columns_to_drop = df_cleaned.columns[df_cleaned.isnull().sum() > threshold]

df_cleaned.drop(columns=columns_to_drop, inplace=True)

df_cleaned.head()
```

Đối với những cột còn lại, thông thường xu hướng chỉ số của một quốc gia tương tự nhau ở những năm liền kề, thế nên, các giá trị thiếu đó sẽ được điền bằng giá trị của năm trước hoặc năm sau nếu như dữ liệu thiếu vẫn chưa được xử lý hết.

```python

df_cleaned = df_cleaned.sort_values(by=['Năm'])

df_cleaned.loc[:, df_cleaned.columns != 'Quốc gia'] = df_cleaned.groupby('Quốc gia').transform('ffill')

df_cleaned = df_cleaned.sort_index()

df_cleaned.head()
```

```python
print("Tỉ lệ % số lượng giá trị thiếu ở mỗi cột sau khi xử lý lần 1: ")

(df_cleaned.isnull().sum()/df_cleaned.shape[0] * 100).round(3)
```

```python
df_cleaned = df_cleaned.sort_values(by=['Năm'])

df_cleaned.loc[:, df_cleaned.columns != 'Quốc gia'] = df_cleaned.groupby('Quốc gia').transform('bfill')

df_cleaned = df_cleaned.sort_index()

df_cleaned.head()
```

```python
print("Tỉ lệ % số lượng giá trị thiếu ở mỗi cột sau khi xử lý lần 2: ")

(df_cleaned.isnull().sum()/df_cleaned.shape[0] * 100).round(3)
```

Sau khi đã thực hiện 2 lần xử lý, ta thấy vẫn còn số lượng nhỏ các cột còn thiếu, thì những giá trị thiếu đó ta sẽ điền bằng giá trị mean của mỗi cột

```python
df_cleaned = df_cleaned.apply(lambda x: x.fillna(x.mean()) if x.name != 'Quốc gia' else x)

df_cleaned.head()
```

```python
print("Tỉ lệ % số lượng giá trị thiếu ở mỗi cột sau khi xử lý lần 3: ")

(df_cleaned.isnull().sum()/df_cleaned.shape[0] * 100).round(3)
```

```python
def plot_distribution(data, title, ax):
    ax.hist(data, bins='auto', alpha=0.7, edgecolor='black')

    ax.set_title(title, fontsize=10, pad=1)
    ax.set_xlabel('Giá trị', fontsize=8)
    ax.set_ylabel('Tần suất', fontsize=8)

    ax.grid(axis='y', linestyle='--', alpha=0.4)
    ax.tick_params(axis='both', which='major', labelsize=8)
```

```python
numeric_cols = df_simplified.select_dtypes('number')

n_cols = 3
n_rows = 8

fig, axes = plt.subplots(n_rows, 3, figsize=(15, n_rows*3))
fig.suptitle('Phân phối các thuộc tính số', y=1.02, fontsize=14)
axes = axes.flatten()


for i, col in enumerate(numeric_cols.columns):
    plot_distribution(df_simplified[col], title=f"{col}", ax=axes[i])

for idx in range(len(numeric_cols.columns), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
def plot_boxplot(data, title, ax):
    ax.boxplot(data,
               flierprops=dict(marker='o', markersize=3, markerfacecolor='#fdae6b'),
               widths=0.6)

    ax.set_title(title, fontsize=10, pad=1)
    ax.set_ylabel('Giá trị', fontsize=8)

    plt.subplots_adjust(wspace=0.3)
```

```python
fig, axes = plt.subplots(n_rows, 3, figsize=(15, n_rows*3.5))
fig.suptitle('Phân bố các thuộc tính số', y=1.02, fontsize=14)
axes = axes.flatten()


for i, col in enumerate(numeric_cols.columns):
    plot_boxplot(df_simplified[col], title=f"{col}", ax=axes[i])

for idx in range(len(numeric_cols.columns), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```
